####  1) Read data and import them into a data.frame or tibble structure

```r
#Import tidyverse packages
library(tidyverse)
library(patchwork)
```

```r
#Filenames and company names
filenames <- c("american_airline_empl.txt","delta_airline_empl.txt","federal_express_empl.txt","united_airline_empl.txt")
companies <- c("AmericanAirline", "DeltaAirline", "FedEx", "UnitedAirline")
```

```r
#Read files
am_air <- as_tibble(read_delim(filenames[1],show_col_types = FALSE))
del_air <- as_tibble(read_delim(filenames[2],show_col_types = FALSE))
fed_ex <- as_tibble(read_delim(filenames[3],show_col_types = FALSE))
un_air <- as_tibble(read_delim(filenames[4],show_col_types = FALSE))
```

####  2) Merge the four data tibble in a common tibble

```r
# Add to each tibble a column with the company's name, then perform inner-join
am_air<- am_air %>% add_column(CompanyName = rep(companies[[1]], nrow(am_air)))
del_air <- del_air %>% add_column(CompanyName = rep(companies[[2]], nrow(am_air)))
fed_ex <- fed_ex %>% add_column(CompanyName = rep(companies[[3]], nrow(am_air)))
un_air <- un_air %>% add_column(CompanyName = rep(companies[[4]], nrow(am_air)))

#Merge using innerjoin. Now total contains the complete tibble
total <- rbind(am_air,del_air,fed_ex,un_air)

#Show the head of the total tibble as an example
head(total)
```

#### 4) - Find the airline companies offering the largest two numbers of flights per day and per week

##### Per day:

```r
#Count flights per day and per company
airlines_per_day <- count(flights, carrier, date) 

#Function to find two carriers with most flights --> returns c(second_to_max, max)
find_two_max<-function(x){
    indices <- order(x$n)
    return(tail(x$carrier[indices],2))
}

#Apply find_two_max function, grouping rows by date
max_per_day <- airlines_per_day %>% group_by(date) %>% do(data.frame(maxCarrier2=find_two_max(.))) 

#Tidy up data
max_per_day <- max_per_day %>% add_column(maxCarrier = c(tail(max_per_day$maxCarrier2,length(max_per_day$maxCarrier2)-1), head(max_per_day$maxCarrier2,1))) %>% distinct(.,date, .keep_all=TRUE)

#Add columns with number of flights, for future plotting
max_per_day <- max_per_day %>% add_column(max = aggregate(airlines_per_day$n, list(airlines_per_day$date), max)$x)
max_per_day <- max_per_day %>% add_column(max2 = aggregate(airlines_per_day$n, list(airlines_per_day$date), function(x) max(x[-order(x)[length(x)]]))$x)
max_per_day <- max_per_day %>% relocate(max2, .after = maxCarrier2)
```

```r
#Show first rows as an example
head(max_per_day,15)
```

```r
#Scatter plot
p <- ggplot()
scatter_max <- geom_point(max_per_day, mapping = aes(x = date, y = max, color = maxCarrier))
scatter_max2 <- geom_point(max_per_day, mapping = aes(x = date, y = max2, color = maxCarrier2))
p <- p + scatter_max + scatter_max2
labels <- labs(x = "Date", y = "N. of flights", colour = "Airline")
p <- p + ggtitle("Two airlines with most flights in 2013") + labels + themes

print(p)
```
