#### Assuming a uniform prior, compute and plot the posterior distribution for Î», the death rate over the measurement time. Determine the posterior mean, median and variance, and compute the 95% credibility interval.

```r
#Observations
deaths_1 <- c(109, 65, 22, 3, 1, 0)
deaths_2 <- c(144, 91, 32, 11, 2, 0)

df <- data.frame(data1 = deaths_1, data2 = deaths_2)
```

```r
#For each of the two samples compute the posterior 
 
results <- data.frame(matrix(nrow = 0, ncol = 5))
colnames(results) <- c("Mean", "Var", "Median", "CI_lower", "CI_upper")
corps <- c("n1", "n2")

par(mfrow = c(1, 2))  # Plotting space
colors <- c("firebrick", "blue")
for(i in 1:ncol(df)){
    #Posterior will be a Gamma function -> find parameters
    alpha <- 1 + sum(df[,i] * seq(from = 0, to = 5, by = 1))
    lambda <- sum(df[,i])

    #Evaluate mean, median, variance
    mean <- integrate(function(x) dgamma(x, shape = alpha, scale = 1/lambda)*x,0,Inf)$value
    median <- qgamma(0.5, shape = alpha, scale = 1/lambda)
    variance <- integrate(function(x) dgamma(x, shape = alpha, scale = 1/lambda)*(x-mean)^2, 0, Inf)$value
                          
    #Credibility interval
    CI <- c(qgamma(0.025, shape = alpha, scale = 1/lambda), qgamma(0.975, shape = alpha, scale = 1/lambda))
    
    #Store values
    results[nrow(results) + 1, ] <- c(mean, variance, median, CI[1], CI[2])                   
                          
    #Plots 
    xs <- seq(from = 0, to = 1, length.out = 1001)        # x points
    yplot <- dgamma(xs, shape = alpha, scale = 1/lambda)  # y points
                          
    plot(xs, yplot, type = "l", main = paste("Posterior after uniform prior, ", corps[i]),
        col = colors[i], lwd = 2, xlab = "\u03bb", ylab = "p(\u03bb)", cex.lab = 1.3, cex.main = 1) #curve
                          
    segments(x0 = mean, y0 = 0, x1 = mean, y1 = dgamma(mean, shape = alpha, scale = 1/lambda),
             col="black", lty = 2, lwd = 2) #mean
                          
    color = col2rgb(colors[i], alpha = FALSE)[,1]
    polygon(c(CI[2], CI[2], CI[1], CI[1], xs[xs > CI[1] & xs < CI[2]]),
            c(dgamma(CI[2], shape = alpha, scale = 1/lambda),0, 0, dgamma(CI[1], shape = alpha, scale = 1/lambda), yplot[xs > CI[1] & xs < CI[2]]), 
            col =rgb(color[1]/255, color[2]/255, color[3]/255, alpha = 0.2))    #shade CI under curve

    #Legend
    legend("topleft", legend = "95% CI",
           fill = rgb(color[1]/255, color[2]/255, color[3]/255, alpha = 0.2) )

}
```

```r
#Display results
rownames(results) <- c("n1", "n2")
results
```

#### A study on water quality of streams, a high level of bacteria X was defined as a level greater than 100 per 100 ml of stream water. n = 116 samples were taken from streams having a high environmental impact on pandas. Out of these, y = 11 had a high bacter X level. indicating with p the probability that a sample of water taken from the stream has a high bacter X level

#### a) Find a frequentist estimator for $p$

```r
n <- 116 #number of samples
y <- 11 #number of samples with high level of bacterium

p <- y/n #Frequentist estimator
print(paste("Frequentist estimator: y/n = ", p))
```

#### b) Using a Beta(1, 10) prior for p, calculate and posterior distribution P (p|y)

```r
# P(p|y) proportional to P(y|p)P(p)
posterior <- function(x){
    return(dbeta(x,1,10) * x^(y) * (1-x)^(n-y) ) # P(y|p) proportional to p^y (1-p)^(n-y)
}
posterior <- Vectorize(posterior)
Z_posterior <- integrate(posterior, 0, 1)$value
```

#### (c) Find the bayesian estimator for p, the posterior mean and variance, and a 95% credible interval

```r
#A bayesian estimator for p is given by the mean of the posterior distribution, E[p].

mean <- integrate(function(x) x*posterior(x)/Z_posterior, 0, 1)$value
var <- integrate(function(x) x^2 * posterior(x)/Z_posterior, 0, 1)$value - mean^2

#We know from the theory that the posterior can be expressed as Beta(1 + y, 10 + n â y)
CI <- c(0.025 %>% qbeta(1+y,10+n-y), 0.975 %>% qbeta(1+y,10+n-y))
```

```r
results <- data.frame(mean, var, CI[1], CI[2])
colnames(results) <- c("Mean", "Var", "CI_lower", "CI_upper")

results
```

```r
#Plot
xs = seq(from = 0, to = 1, length.out = 1001)
df = data.frame(xs, ys = dbeta(x = xs, 1+y, 10+n-y))

p <- ggplot(df) + geom_line(aes(x = xs, y = ys, color = "Posterior")) + ggtitle("Posterior, Beta prior") + theme
p <- p + geom_segment(aes(x = mean, y = 0, xend = mean, yend = 15, color = "Mean"), linetype = "dashed")
p <- p + geom_ribbon(data=subset(df ,xs>=CI[1] & xs<=CI[2]),aes(x = xs,ymax=ys, fill="Confidence Interval"),ymin=0,colour=NA,alpha=0.2)
p <- p + labs(x = "p", y = "P(p|y)", color = NULL, fill = NULL)

options(repr.plot.width=8, repr.plot.height=6)
p + sc + xlim(0,0.25)
```

#### d) Test the hypotesis H0 : $p = 0.1$ versus H1 : $p\neq0.1$ at 5% level of significance with both the frequentist and bayesian approach

##### Frequentist approach
$H_0: p = 0.1$
$H_1: p!= 0.1$ 
We measure $y$ contaminated samples. If $H_0$ holds true, then the distribution of $y$ will be a binomial with $p=0.1$.

```r
#I write it as a function, because I will also need it later in point h.
frequentist_approach <- function(y_func,n_func, color){
    xs <- seq(from = 0, to = n_func, by = 1)
    values <- dbinom(xs, size = n_func, prob = 0.1) #binomial distribution
    df <- data.frame(xs, values)

    #Two sided --> find approximately 0.025 on one side and 0.975 on the other
    cumulatives <- rep(0, length(values)) 
    cumulatives[1] = values[1]
    for(i in 2:length(values)){
        cumulatives[i] <- cumulatives[i-1] + values[i]
    }
    cumulatives <- unlist(cumulatives)

    ##Find first entry in cumulatives >= 0.025, and check the previous value. Choose the value closest to 0.025
    indeces_min <- c(min(which(cumulatives >= 0.025))-1, min(which(cumulatives >= 0.025)))
    indeces_max <- c(max(which(cumulatives <= 0.975)), max(which(cumulatives <= 0.975))+1)
    
    if(abs(cumulatives[indeces_min[1]] - 0.025) < abs(cumulatives[indeces_min[2]]-0.025)){
        indeces_min <- indeces_min[1]
    }
    else{
        indeces_min <- indeces_min[2]
    }
    
    if(abs(cumulatives[indeces_max[1]] - 0.975) < abs(cumulatives[indeces_max[2]]-0.975)){
        indeces_max <- indeces_max[1]
    }
    else{
        indeces_max <- indeces_max[2]
    }
    
    x_min <- xs[indeces_min]
    x_max <- xs[indeces_max]

    alpha_used = cumulatives[indeces_min] + (1 - cumulatives[indeces_max])
    
    print(paste("First point in acceptance region: ", x_min))
    print(paste("Last point in acceptance region: ", x_max))
    
    print(paste("Alpha used: ", alpha_used))
    
    print(paste("Lower tail integral: ", cumulatives[indeces_min]))
    print(paste("Upper tail integral: ", cumulatives[indeces_max]))
    
    #Plot and show where our observation lands. Change colour to highlight acceptance region
    x_barplot = seq(from = 0, to = 30, by = 1)
    
    colours = rep("white", length(x_barplot))                # only plot from 0 to x_max + 5 to "zoom in"
    colours[x_barplot >= x_min & x_barplot <= x_max] = color # acceptance region
    
    barplot(dbinom(x_barplot, n_func, 0.1), col = colours, 
            names.arg = seq(from = 0, to = x_barplot[length(x_barplot)], by = 1), 
            main = paste("Binomial distribution with p = 0.1, n = ", n_func, ", observation = ", y_func),
           xlab="y", ylab="P(y|p = 0.1)", cex.lab = 1.2, las=2) #binomial, restricted to [0,35] for plotting
    
    #Add legend
    legend("topright", legend = c("Acceptance region", "Rejection region"), fill = c(color, "white") )
    
}
```

```r
options(repr.plot.width=10, repr.plot.height=6)

frequentist_approach(y,n,"lightcoral")
```

The observation $y=11$ is well inside the acceptance region, so we don't reject the null hypothesis.
