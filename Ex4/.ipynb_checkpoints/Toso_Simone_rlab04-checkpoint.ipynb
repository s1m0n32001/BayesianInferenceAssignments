### Exercise 1
The number of claims received by an insurance company during a week follows a Poisson distribution with unknown mean (μ).

The number of claims, per week, observed ovr a ten week period are:
5, 8, 4, 6, 11, 6, 6, 5, 6, 4

(a) suppose to use a prior uniform distribution for μ
- find the posterior distribution for μ and compute the posterior mean, median and variance
- plot the posterior distribution and the 95% credibility interval

#### Find the posterior distribution for μ and compute the posterior mean, median and variance

```r
claims <- c(5,8,4,6,11,6,6,5,6,4)

#The posterior distribution will be a Gamma function as seen in class
posterior_uniform <- function(x){
    return(dgamma(x, shape = sum(claims)+1, scale =1/length(claims)))
}

#Evaluate mean, median, variance
mean_uniform <- integrate(function(x) x*posterior_uniform(x), 0, 20)$value
median_uniform <- qgamma(0.5, shape = sum(claims)+1, scale =1/length(claims))
variance_uniform <- integrate(function(x) (x-mean_uniform)^2*posterior_uniform(x),0,20)$value
                  
print(paste("Mean",mean_uniform))
print(paste("Median",median_uniform))
print(paste("Variance", variance_uniform))
```

#### Plot the posterior distribution and the 95% credibility interval

```r
#Compute credibility interval
CI_uniform <- c(qgamma(0.025, shape = sum(claims)+1, scale =1/length(claims)), qgamma(0.975, shape = sum(claims)+1, scale =1/length(claims)))

#Make plot
xs <- seq(from = 0, to = 20, by = 0.1)
p <- ggplot() + geom_line(aes(x = xs, y = posterior_uniform(xs), color = "Posterior")) + labs(x = "\u03bc", y = "p(\u03bc)", color = NULL)
p <- p + geom_segment(aes(x = CI_uniform[1], xend = CI_uniform[1], y = 0, yend = 0.5, color = "CI"), linetype = "dashed")
p <- p + geom_segment(aes(x = CI_uniform[2], xend = CI_uniform[2], y = 0, yend = 0.5, color = "CI"), linetype = "dashed")

sc <- scale_colour_manual(values = c("firebrick","black"))
p + ggtitle("Posterior distribution, uniform prior") + theme + sc
```

#### (b) Suppose to use a Jeffreys’ prior for $\mu$  $g(μ) \propto 1/\sqrtμ$
    
#### Find the posterior distribution for μ and compute the posterior mean, median and variance


```r
#Posterior after Jeffrey's prior
posterior_jeffrey <- function(x){
    return(1/sqrt(x) * dgamma(x, shape = sum(claims)+1, scale =1/length(claims)))
}
posterior_jeffrey <- Vectorize(posterior_jeffrey)

#Integrate on [0, +inf] so that later we can normalize it
Z_posterior_jeffrey <- integrate(posterior_jeffrey, 0, Inf)$value

#Cumulative function
cumulative <- function(x){
    return(integrate(posterior_jeffrey,0,x)$value/Z_posterior_jeffrey)
}

#Evaluate mean, variance, median. The latter is done by inverting the cumulative function
mean_jeffrey <- integrate(function(x) x*posterior_jeffrey(x)/Z_posterior_jeffrey, 0, Inf)$value
variance_jeffrey <- integrate(function(x) (x-mean_jeffrey)^2*posterior_jeffrey(x)/Z_posterior_jeffrey, 0, Inf)$value
median_jeffrey <- inverse(cumulative, lower = 0.0001, upper = 1000)(0.5)                  

print(paste("Mean",mean_jeffrey))
print(paste("Median",median_jeffrey))
print(paste("Variance", variance_jeffrey))
```

#### Plot the posterior distribution and the 95% credibility interval

```r
#Compute CI
xs <- seq(from = 0.1, to = 20, by = 0.1)
CI_jeffrey <- c(inverse(cumulative, lower = 0.0001, upper = 1000)(0.025), inverse(cumulative, lower = 0.0001, upper = 1000)(0.975))

p <- ggplot() + geom_line(aes(x = xs, y = posterior_jeffrey(xs)/Z_posterior_jeffrey, color = "Posterior")) + labs(x = "\u03bc", y = "p(\u03bc)", color = NULL)
p <- p + geom_segment(aes(x = CI_jeffrey[1], xend = CI_jeffrey[1], y = 0, yend = 0.5, color = "CI"), linetype = "dashed")
p <- p + geom_segment(aes(x = CI_jeffrey[2], xend = CI_jeffrey[2], y = 0, yend = 0.5, color = "CI"), linetype = "dashed")

sc <- scale_colour_manual(values = c("firebrick","black"))
p + ggtitle("Posterior distribution, Jeffrey's prior") + theme + sc
```

#### c) Evaluate a 95% credibility interval for the results obtained with both priors. Compare the result with that obtained using a normal approximation for the posterior distribution, with the same mean and standard deviation

```r
#Use qnorm for simplicity
CI_uniform_norm <- c(qnorm(0.025, mean_uniform, sqrt(variance_uniform)), qnorm(0.975, mean_uniform, sqrt(variance_uniform)))
CI_jeffrey_norm <- c(qnorm(0.025, mean_jeffrey, sqrt(variance_jeffrey)), qnorm(0.975, mean_jeffrey, sqrt(variance_jeffrey)))

ConfidenceIntervals <- t(data.frame(CI_uniform, CI_uniform_norm, CI_jeffrey, CI_jeffrey_norm))
colnames(ConfidenceIntervals) <- c("Lower bound", "Upper bound")
```

### Exercise 2

A well established and diffused method for detecting a disease in blood fails to detect the presence
of disease in 15% of the patients that actually have the disease.

A young UniPD startUp has developed an innovative method of screening. During the qualification
phase, a random sample of n = 75 patients known to have the disease is screened using the new
method.

#### a) What is the probability distribution of y, the number of times the new method fails to detect the disease ?
#### Answer: 
The probability distribution of y will be a binomial distribution. If $p$ is the probability of the new method to correctly detect the disease in an ill person, we have $p(y) = \binom{n}{y} p^y (1-p)^{n-y}$.

#### b) On the n = 75 patients sample, the new method fails to detect the disease in y = 6 cases. What is the frequentist estimator of the failure probability of the new method ?
#### Answer:
An unbiased frequentist estimator for the parameter $p$ is $\hat{p}=\frac{y}{n}$. It is unbiased because:

$E[\hat{p}] = \frac{1}{n}E[y] = \frac{1}{n} \sum_{y} \binom{n}{y} p^y (1-p)^{n-y} y = \frac{1}{n}\frac{d}{dx}\sum_{y}\binom{n}{y} p^y (1-p)^{n-y}x^y |_{x=1} = \frac{1}{n}\frac{d}{dx}(1-p+px)^{n}|_{x=1} = p$.  

#### c) Setup a bayesian computation of the posterior probability, assuming a beta distribution with mean value 0.15 and standard deviation 0.14. Plot the posterior distribution for y, and mark on the plot the mean value and variance.

```r
#alpha, beta parameters of Beta --> mean = alpha/(alpha + beta), var = alpha*beta/((alpha+beta)^2 * (alpha + beta + 1))
m <- 0.15
s <- 0.14

y <- 6
n <- 75

xs <- seq(from = 0.001, to = 1, length.out = 1000)


#Compute alpha and beta for Beta distribution
alpha <- m*(m^2*(1/m-1)/s^2-1)
beta <- (1/m-1)*alpha

#Evaluate betas on xs, for future plotting
betas <- dbeta(xs,alpha,beta) 

#Posterior distribution
posterior <- function(x){
    return(dbeta(x,alpha,beta)*(x)^y*(1-x)^(n-y))
}

posterior <- Vectorize(posterior)
#Integral over [0,1] for normalization
Z_posterior <- integrate(posterior,0,1)$value

#Mean value
mean <- integrate(function (x) x*posterior(x), 0, 1)$value / Z_posterior

#Variance
var <- integrate(function(x) (x-mean)^2*posterior(x),0,1)$value / Z_posterior
p <- ggplot()                 
p <- p + geom_line(aes(x=xs, y = posterior(xs)/Z_posterior, color = "Posterior"))
p <- p + geom_segment(aes(x = mean-sqrt(var), y = 0, xend =  mean-sqrt(var), yend = 14, color = "Mean +- std"), linetype = "dashed") + geom_segment(aes(x = mean+sqrt(var), y = 0, xend =  mean+sqrt(var), yend = 14, color = "Mean +- std"), linetype = "dashed") 
p <- p + geom_segment(aes(x = mean, y = 0, xend = mean, yend = 14, color = "Mean value"))
                 
sc <- scale_colour_manual(values = c("firebrick","blue","black", "purple"))
```

```r
p + theme + ggtitle("Posterior, mean and std") + labs(x = "\u03c0", y = "p(\u03c0)", color = NULL) + sc
```

#### d) Perform a test of hypothesis assuming that if the probability of failing to the detect the desease in ill patients is greater or equal than 15%, the new test is no better that the traditional method. Test the sample at a 5% level of significance in the Bayesian way.

```r
#Take the posterior and integrate it from 0 to 0.15 
integral <- integrate(posterior, 0, 0.15)$value / Z_posterior
pValue <- 1-integral
print(paste("Prob. failure rate > 0.15 = ", pValue))
```

We obtain a $3.1\%$ probability that $p > 0.15$. Therefore we reject ($\alpha = 5\%$) the hypothesis that the failure probability is $>15\%$.

#### e) Perform the same hypothesis test in the classical frequentist way.

```r
#Evaluate probability to obtain 6 or less failures if prob = 0.15. If < 5%, then we reject the null hypothesis H0: p>=0.15.
ys <- seq(0,y,by = 1)
binoms <- list()
for(i in 1:length(ys)){
    binoms[i] <- dbinom(ys[i],n,prob = 0.15)
}
binoms <- unlist(binoms)
total <- sum(binoms)
print(total)
```

If $p = 0.15$, the probability to obtain the measured data ($y = 6$ failures over $n = 75$ patients) is $0.054 > 0.05$, so we cannot reject the hypothesis that the failure probability is $> 15\%$.

### Exercise 3
Lighthouse problem - but this time we don't know neither $\alpha$ nor $\beta$.

#### Solution
As done in class, we can derive the $p(x)$ from the $p(\theta)$, which is done through a change of variables. We get
$p(x)dx = p(\theta)d\theta \implies \frac{1}{\pi}d\theta = p(x) |\frac{dx}{d\theta}|d\theta$. We have $x = \alpha + \beta \tan{\theta}$ which leads to $p(x) = \frac{1}{\pi\beta}\frac{1}{1+\left(\frac{x-\alpha}{\beta}\right)^2}$.

We can evaluate the posterior $p(\alpha,\beta|D) \propto p(D|\alpha,\beta)p(\alpha,\beta)$. To study this scenario, we can generate data having $\alpha_{true}=1\rm{km}, \beta_{true} = 1.5\rm{km}$. The prior we choose is uniform in the $[-2\rm{km}, 2\rm{km}] \times [1\rm{km}, 3\rm{km}]$ domain.

##### Generate data

```r
#True values
alpha_true <- 1
beta_true <- 1.5

#Save data in list. Take N samples
samples <- list()
N <- 10000

for(i in 1:N){
    theta <- runif(1)*pi
    samples[i] = alpha_true + beta_true * tan(theta)
}
samples <- as.numeric(unlist(samples))


#Theorical distribution
cauchy <- function(x){
    return(1/(pi*beta_true)*1/(1+((x-alpha_true)/(beta_true))^2))
}
cauchy <- Vectorize(cauchy)

#Plot theoretical distribution and simulated data
samplesPlot <- samples[(samples > -20) & (samples < 20)] 
xs <- seq(from = -20, to = 20, length.out = 2001) 
plot <- ggplot() + geom_density(aes(x = samplesPlot, fill = "Density"), alpha = 0.1) 
plot <- plot + geom_line(aes(x=xs, y = cauchy(xs), color = "Theoretical distribution"), linewidth = 1)
plot <- plot + ggtitle("Sampled distribution and theoretical prediction")
plot + theme + labs(x = "x", y ="p(x)", fill = NULL, color = NULL)
```

##### Evaluate posterior
Let's use the sampled data to find a posterior distribution for the parameters $\alpha$ and $\beta$.

```r
# Let's represent alpha and beta on a grid and compute the posterior
## Grid [-2,2] x [1,3] --> use a grid step of 0.1 x 0.1 --> area = 0.01
y <- seq(from = -2, to = 1.9, by = 0.1)
x <- seq(from = 1, to = 2.9, by = 0.1)
data <- expand.grid(X=x, Y=y)
grid <- matrix(rep(1/8,40*20), nrow = 40, ncol = 20)
```

```r
#We evaluate the posterior sequentially at each "timestep" (i.e. number of measurements).
## Make a dataframe in which we will store the (alpha, beta, p(alpha, beta), time) rows.
z <- list()
for(i in 1:nrow(data))
    {
    row <- as.integer(((data[i,]$Y+2) / 0.1)) + 1
    col <- as.integer(((data[i,]$X-1)/0.1)) + 1
    z[i] <- grid[row,col]
}
z <- as.numeric(unlist(z))
data <- data.frame(X = data$X,Y = data$Y, z)

nLines <- nrow(data)
time <- rep(0,nLines)

data$"time" <- time

dataframe <- data
```

```r
#Compute posterior for later timesteps and add rows to dataframe.
## To speed it up we only store the posterior at certain times.
counter <- 1
frames <- c(1,2,3,4,5,10,20,30,40,50,100,200,300,400,500,1000,5000,10000)
for(sample in samples){
    for(row in 1:nrow(grid)){
        for(col in 1:ncol(grid)){
            alpha <- 2 - (row-1)*0.1 
            beta <- 1 + (col-1)*0.1 
            grid[row,col] <- grid[row, col] * (1/(pi*beta)*1/(1+((sample-alpha)/(beta))^2)) #Update posterior
        }
    }
    
    #Normalization
    tot <- sum(grid*0.01)
    grid <- grid/tot
    
    #Store grid
    if(any(counter==frames)){
        for(row in 1:nrow(grid)){
            for(col in 1:ncol(grid)){
                alpha <- 2 - (row-1)*0.1
                beta <- 1 + (col-1)*0.1
                new_row <- c(beta, alpha, grid[row,col], counter)
                dataframe[nrow(dataframe)+1, ] <- new_row
            }
        }
    }
    counter <- counter + 1
}
```

Let's plot a heatmap of the posterior at various stages of the sampling. The colormap indicates the probability of $(\alpha,\beta)$ being in that square of the grid.

```r
labels <- labs(x = "beta [km]", y = "alpha [km]", fill = "p(alpha,beta)")
theme <- theme(legend.text = element_text(size = 10), axis.title = element_text(size = 10), axis.text = element_text(size = 10), plot.title = element_text(face = "bold", size = 10))

#Make plots in for cycle
plot_list <- list()
counter <- 1
for(frame in frames){
    p <- ggplot(dataframe[dataframe$time == frame, ], aes(x = X, y = Y, fill = z*0.01)) + geom_tile() + ggtitle(paste("Posterior, nSamples = ", frame)) + labels + theme
    plot_list[[counter]] <- p
    counter <- counter + 1
}

#Display some plots
options(repr.plot.width=8, repr.plot.height=12)

p <- ggarrange(plotlist = plot_list[c(1,5,6,10,11,15,16,18)], widths = rep(5,8), ncol = 2, nrow = 4)
p
```

As shown in the heatmaps, as the number of samples increases the posterior gets more and more peaked around $(\alpha_{true},\beta_{true})$.

### Exercise 4
(a) vary the sampling resolution used to generate the data, keeping the same sampling range
xdat <- seq(from=-7*w, to=7*w, by=0.5*w)
 - Change the resolution w = {0.1, 0.25, 1, 2, 3}
 - Check the effect on the results

```r
#Keep sampling range in [-7,7] and increase resolution
ws <- c(0.1,0.25,1,2,3)

#Parameters
a_true <- 2
b_true <- 1
x0 <- 0
t <- 5
w <- 1

#Signal function
signal <- function(x,a,b,w){
    return(t * (a*exp(-(x-x0)^2/(2*w^2)) + b))
}

#Dataframe in which we'll store mean and std for a and b
df = data.frame(matrix(nrow = 0, ncol = 5))
colnames(df) <- c("w","E[A]", "Var[A]", "E[B]", "Var[B]")

set.seed(12345)

#Plots
par(mfrow= c(5, 3))
options(repr.plot.width=10, repr.plot.height=15)

#For each value of w check the posterior 
for(w_res in ws){
    #Positions on x axis with this resolution
    xdat <- seq(from = -7, to = 7, by = 0.5*w_res)
    
    #Simulate data taking
    expectations <- signal(xdat, a_true, b_true, w)     #Expected signal
    simulated_data <- rpois(n = length(xdat), lambda = expectations)  #Simulated counts
    
    #Now evaluate posterior on a grid
    as <- seq(from = 1, to = 3, by = 0.01)
    bs <- seq(from = 0.1, to = 2, by = 0.01) # 0.1 as increment, so each "square" of the domain will have area 0.01
    grid <- matrix(rep(1, length(as)*length(bs)), nrow = length(as), ncol = length(bs)) #we'll normalize later
    
    #We use a uniform prior, so we simply evaluate p(Data|a,b)
    for(row in 1:nrow(grid)){
        for(col in 1:ncol(grid)){
            a <- 1 + (row-1)*0.01
            b <- (col-1)*0.01  
            grid[row, col] <- sum(dpois(simulated_data, lambda = signal(xdat, a, b, w), log = TRUE))
        }
    }
    
    #Normalize grid
    grid <- grid - max(grid, is.na = FALSE)
    grid <- exp(grid)
    grid <- grid/(sum(grid)*0.01^2)

    #Plot contour
    contour (as, bs, grid, nlevels = 4, labcex = 1, lwd = 1, xlab="A", ylab="B", main=paste("Posterior, w = ", w_res), ylim = c(0.8,1.4), cex.main=2, cex.lab=1.5)
    abline (v=2,h=1,col="black", lty = 2)
    
    #Evaluate mean and std
    p_a <- apply(grid, 1, sum)
    p_a <- p_a/sum(p_a*0.01)
    mean_a <- sum(p_a*as)/sum(p_a)
    var_a <- sum(p_a*(as-mean_a)^2)/sum(p_a)
    
    p_b <- apply(grid, 2, sum)
    p_b <- p_b/sum(p_b*0.01)
    mean_b <- sum(p_b*bs)/sum(p_b)
    var_b <- sum(p_b*(bs-mean_b)^2)/sum(p_b)

    #Plot marginalized distributions
    plot(as,p_a, type = "l", xlab = "A", ylab = "p(A) marginalized", main = paste("p(A|D), w = ", w_res), cex.main=2, cex.lab=1.5)
    abline (v=a_true, col="black", lty = 2)

    plot(bs, p_b, type = "l", xlab = "B", ylab = "p(B) marginalized", main = paste("p(B|D), w = ", w_res), cex.main=2, cex.lab=1.5)
    abline (v=b_true, col="black", lty = 2)

    df[nrow(df)+1,] <- c(w_res, mean_a, var_a, mean_b, var_b) 

}
```

#### b ) Change the ratio A/B used to simulate the data (keeping both positive in accordance with the prior)
 - Check the effect on the results

```r
#Sampling range
xdat <- seq(from = -7, to = 7, by = 0.5)  

#Parameters
a_true <- 2
w <- 1
x0 <- 0
t <- 5

#Signal function
signal <- function(x,a,b,w){
    return(t * (a*exp(-(x-x0)^2/(2*w^2)) + b))
}

#Dataframe in which we'll store mean and std for a and b
df = data.frame(matrix(nrow = 0, ncol = 5))
colnames(df) <- c("w","E[A]", "Var[A]", "E[B]", "Var[B]")

set.seed(12342)
#For each value of B/A check the posterior posterior
AB_ratios <- c(0.1, 0.25, 0.5, 1., 1.5, 1.8)

#Plot parameters
par(mfrow= c(2, 3))
options(repr.plot.width=13, repr.plot.height=13)

for(ratio in AB_ratios){
    
    b_true_ratio <- a_true*ratio
    
    #Simulate data taking
    expectations <- signal(xdat, a_true, b_true_ratio, w)     #Expected signal
    simulated_data <- rpois(n = length(xdat), lambda = expectations)  #Simulated counts
    
    #Now evaluate posterior on a grid
    as <- seq(from = 1, to = 3, by = 0.01)
    bs <- seq(from = 0.1, to = 2, by = 0.01) # 0.1 as increment, so each "square" of the domain will have area 0.01
    grid <- matrix(rep(1, length(as)*length(bs)), nrow = length(as), ncol = length(bs)) #we'll normalize later
    
    #We use a uniform prior, so we simply evaluate p(Data|a,b)
    for(row in 1:nrow(grid)){
        for(col in 1:ncol(grid)){
            a <- 1 + (row-1)*0.01
            b <- (col-1)*0.01  
            grid[row, col] <- sum(dpois(simulated_data, lambda = signal(xdat, a, b, w), log = TRUE))
        }
    }
    
    #Normalize grid
    grid <- grid - max(grid, is.na = FALSE)
    grid <- exp(grid)
    grid <- grid/(sum(grid)*0.01^2)
    
    #Evaluate mean and std
    p_a <- apply(grid, 1, sum)
    mean_a <- sum(p_a*as)/sum(p_a)
    var_a <- sum(p_a*(as-mean_a)^2)/sum(p_a)
    
    p_b <- apply(grid, 2, sum)
    mean_b <- sum(p_b*bs)/sum(p_b)
    var_b <- sum(p_b*(bs-mean_b)^2)/sum(p_b)

    
    df[nrow(df)+1,] <- c(ratio, mean_a, var_a, mean_b, var_b)   
    
    xdat.off <- xdat -0.25 
    x_cont <- seq(from = -7, to = 7, by = 0.001)
    plot(xdat.off , simulated_data, type="s",col="firebrick", lwd=2, main=paste("Posterior, ratio = ", ratio), xlab = "x", ylab = "Counts", cex.main=2, cex.lab=1.5, )
    lines(x_cont, signal(x_cont, mean_a, mean_b, w), col = "black")
    
 
}
```
