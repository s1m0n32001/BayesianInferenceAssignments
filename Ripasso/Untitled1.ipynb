```r
library(rjags)
library(coda)
```

```r
observations <- c(1,0,0,0,0,0,1,1,0,1,0,0,1,0,0,0,1,0,0,0,0,0,1)

data <- NULL
data$X <- observations

model <- jags.model("cassoooo.bugs", data)
```

```r
update(model, 2000)
```

```r
chain <- coda.samples(model, "p", 10000)
```

```r
plot(chain, col = "lightcoral")
```

### Pfizer study

```r
tot_vacc <- 21999
tot_plac <- 21539

pos_vacc <- 8
pos_plac <- 86
```

```r
data <- NULL
vacc_obs <- rep(0, tot_vacc)
vacc_obs[1:pos_vacc] <- 1

data$X <- vacc_obs

model_vacc <- jags.model("vacc.bugs", data)
```

```r
update(model_vacc, 3000)
```

```r
chain <- coda.samples(model_vacc, n.iter = 10000, variable.names = "p")
```

```r
data_plac <- NULL
plac_obs <- rep(0, tot_plac)
plac_obs[1:pos_plac] <- 1

data$X <- plac_obs

model_vacc <- jags.model("vacc.bugs", data)
```

```r
update(model_vacc, 3000)
```

```r
chain_plac <- coda.samples(model_vacc, n.iter = 10000, variable.names = "p")
```

```r
plot(chain)
plot(chain_plac)
```
