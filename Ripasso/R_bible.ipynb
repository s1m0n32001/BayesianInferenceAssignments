### precision limits of floating points

Floating point numbers are stored as doubles (64bit), while numbers and sequences are represented as integers. Only operations among integers are exact. But we can check for near equality!

- **identical()** checks if R objects are exactly equal
- **all.equal()** checks for near equality

````r
u <- sqrt(2)
u * u == 2   # FALSE, but we know it is true!

# also, the default print fails you...
print(u*u)
print(u*u, digits=19) # ...you need more precision
````

````r
identical( u*u, 2) # same as above
all.equal( u*u, 2) 
````

### **Vector attributes**

Attributes are **name & value** pairs attached to an object. Attributes are used to create user-defined S3 classes.

````r
y <- c( 25, 12, 7, 4, 6, 2, 1, 0, 2)

attr( y, "nnxame") <- "attr_value"  # assign an attribute
attributes( y )                     # list all attributes
````

Most attributes are lost during operations, unless they are part of an S3 class. Only *names* and *dim* attributes are preserved.

#### **Names**

A vector can be given a name in three ways:
- when creating it, as flags in c() function
- by assigning a character vector to names()
- inline, using setNames()

````r
x <- c( a = 1, b = 2, c = 3)    # in creation

x <- 1:3
names(x) <- c( "a", "b", "c")   # assigning

x <- setNames( 1:3, c( "a", "b", "c") )  # inline

# to show the names:
names(x)

# to remove the names:
unname(x)         # note: this prints x
names(x) <- NULL  # ... or by assign
````

### **Matrices and arrays**

A matrix is a 2D object, an array is a nD object (n>2).

Matrix elements can be accessed through `m[ ROW, COLUMN]`.

You can apply a function $f$ to a matrix $M$ along rows (1) or columns (2) using `apply(M, 1 or 2, f)`.

````r
# MATRICES =======================

# ex1: specify only the rows
x <- matrix( c(1,0,0,0,1,0,0,0,1), nrow=3); x

class(x); attributes(x); str(x);

# ex2: specify rows and columns
v1 <- c( 1:20 )
matrix( v1, nrow = 4, ncol = 5)
#  or alternatively:  matrix( v1, c(4,5))

# ex3: change the dim attribute
dim(v1) <- c(4,5)
is.matrix(v1)


# ARRAYS =======================
array(v1, c( 2, 5, 2) ) 
  #  set the dimensions from vector
````

````r
# you can create a matrix arranging values
#  by row or column

vct <- 1:8

M <- matrix( vct, byrow=T, nrow=2);  M

M <- matrix( vct, byrow=F, nrow=2);  M
````

#### Add **rows and columns** to a matrix

````r
M <- matrix( 1:8, byrow=T, nrow=2)

# You can add rows and columns...
M <- rbind( M,  colMeans(M) )
M <- cbind( M,  rowMeans(M) )

# ... and set their names, if you like
rownames(M) <- c('pippo', 'pluto', 'col_mean')
colnames(M) <- c('a','b','c','d', 'row_mean')  
M
````

### example: compute a factorial

You can use `factorial(n)`, of course... but also built-in R functions, like `max(cumprod(1:n))`. As an exercise you can do it with a loop.

````r
n <- 5

i <- 1; fac <- 1;
while(i < n) {
    i <- i+1
    fac <- fac*i
}
cat('with while:', fac,'\n')


i <- 1; fac <- 1;
repeat {
    fac <- fac*i
    if( !(i<n) ) break
    i <- i+1
}
cat('with repeat:', fac,'\n')
````

## **vectorialization**

We have already seen that `ifelse()` allows to perform if operations in a vectorial way.

In general, loops are slower than vectorized operations.

*Note*: install *microbechmark* first (`install.packages("microbenchmark")`).

````r
fm1 <- function(x) { max(x) }

fm2 <- function(x) {
    cmax <- x[1]
    for( i in 2:length(x) )
        if(x[i] > cmax) cmax <- x[i]
    }

## let's test them!
library( microbenchmark )

u <- runif(10^2) # generates random values

microbenchmark( max(u), fm1(u), fm2(u) )
````

````r
# example: generate vector from 1 to n

t1 <- function(n) { y <- 1:n }
# this is faster

t2 <- function(n) { 
        y <- numeric(n)
        for(i in 1:n) y[i] <- i
       }
# this function allocates the vector
#  then performs the operations

t3 <- function(n) { 
        y <- NULL
        for(i in 1:n)
            y <- c(y, i)
       }
# this function grows the vector at every iter!
#  -> DON'T DO THIS
````

# **Dataframes and lists**

Lists are evolution of atomic vectors. Each element of a list can be of any type, because lists contain references to other R objects.

````r
l1 <- list(1:3,
           "list element",
           c(TRUE,FALSE,FALSE),
           c(3.5, -1.7) )

typeof(l1)

l2 <- l1
l2[[1]] <- c( 1L, 2L, 4L)
# double enclosure [[ ]]
l2
````

A **dataframe** is:
- a named list of vectors of same length,
- with attributes for columns and rows names (`names`, `row.names`), 
- belonging to the `data.frame` class.

Note: by default, from R 4.0, strings are imported as factors: `stringsAsFactors = FALSE`.

````r
df1 <- data.frame( x=1:3, y=letters[1:3])
typeof(df1)


head(df1, n=2) # extract the first two lines

# accessing elements (columns)...
df1[[1]]   # ... by component index
df1$y      # ... by component name
df1[ , 1]  # ... you can treat them as matrices
````

Because of `copy-on-modify`, if a column is changed, only the reference to the new column will be updated. But if we **change a row**? Every column must be updated!

Modifying an R object usually creates a copy, but there are 2 exceptions:
- objects with single binding, get a special performance optimization
- environments, a special type of object, which are always modified in place
but it is very difficult to predicts when R applies this optimization.

Concerning object binding, R only counts 0, 1 or MANY.
- Example: if an object has 2 bindings (i.e. many), and one gets deleted, the reference does not go back to 1 (many - 1 = many)

When a funtion is called, it makes a reference to the object, so it is very difficult to predict weather or not a copy will occur.

More: https://developer.r-project.org/Refcnt.html

````r
library(lobstr)

v <- c ( 1, 3, 2)
lobstr::obj_addr(v)

v[[3]] <- -2
lobstr::obj_addr(v)
````

# **Subsetting**

*apply(), lapply(), sapply(), tapply() and mapply()* are used to perform operations on all the elements of a complex object avoiding the use of loops.

**Function: apply(X, MARGIN, FUN)** with
- X: an array or matrix
- MARGIN : a value or range between 1 and 2 to define where to apply the function :
    - MARGIN = 1 : the manipulation is performed on rows
    - MARGIN = 2 : the manipulation is performed on columns
    - MARGIN = c ( 1 ,2 ) the manipulation is performed on rows and columns
- FUN : which function to apply.

````r
df <- data.frame(exam1=c(27,28,24,24,30,26),
                 exam2=c(25,30,26,24,30,30),
                 labrep=c(29,30,29,27,29,30))

# mean on column 2 and 3 (exam1 and 2), along rows
row_avg <- apply( df[,1:2], MARGIN=1, FUN=mean)
row_avg

# apply a function to each element
apply( df, 1, function(x) x^2 + x)
````

`lapply(LIST, FUN)` returns a list, computing the function FUN on each list in LIST. 
  
`sapply()` is a simplified version
  - it returns, in general, a vector
  - it is more specific, and less general
  
more specialized functions exist:
  - `vapply()` can specify a return object
  - `mapply()` can use multiple arguments
  - `rapply()` can be used recursively

````r
plants <- list(height=runif(10,min=10,max=20),
               mass=runif(10,min=5,max=10),
               flowers=sample(1:10,10))

lapply(plants, mean) # output is a list

# we can cast it to a simple named vector
v <- unlist(lapply(plants, mean))
v[1]
````

````r
# to apply a function to a vector
sapply( 12:14, seq)  # generate a list of seq, from 1:12 to 1:14
#  equivalent to   seq(12);  seq(13);  seq(14);
````

# **Graphics**

There are several graphics systems in R.

**base graphics**

- a pen on paper model: you can only draw on top of a plot, no modification or deletion of existing content possible
- no user accessible representation of a graphics, only appearance on the screen
- fast primitives, but with limited scope

**grid graphics**
- developed by Paul Murrell (started in his PhD work)
- graphical objects can be represented independently of the plot and modified later
- a system of viewports makes it easier to lay out complex graphics

**lattice graphics**
- developed by Deepayan Sarkar
- use grid graphics to implement the trellis graphics system of Cleveland

**ggplot2**
- a data visualization package creatd by Hadley Wickham in 2005
- it implements L. Wilkinsonâ€™s *Grammar of Graphics*: a general scheme for data visualization which breaks up graphs into semantic components such as scales and layers

## **ggplot2**

**The Grammar of Graphics**

Created by Wilkinson in 2005 to describe the features living behind all statistical graphics.
It has the following components:
- **layer**: are used to create the objects on a plot. They are defined by five basic parts: 
    - **data** (the source of the information to be visualized),
    - **mapping** (how variables are applied to the plot),
    - **statistical transformation** (which transform the data by summarizing the information),
    - **geometric object** (controls the type of the plot to be created),
    - **position adjustment**.
- **scale**: controls how data is mapped to aesthetic attributes (ex: scale for colors)
- **coordinate system**: maps the position of objects onto the plane of the plot and controls how axes and grid lines are drawn
- **faceting**: can be used to split data into subsets of the entire dataset

*Ref:*  Wilkinson L., The grammar of graphics. Statistics and computing 2005, Springer, New York


#### **Plots**

Every plot has three components :
1. **data** to be plotted
2. **aesthetics**, a set of mappings between variables in the data and visual properties
3. **geoms**, a layer that describes how to render each observation
```R
ggplot( df, aes( x=displ, y=hwy) ) + geom_point()
#      (1)     (2)                       (3)
```

Data and aesthetic mappings are supplied in ggplot(), additional layers are added on with the â€™+â€™ operator.

````r
library(ggplot2)

p1 <- ggplot(mpg, aes(displ, hwy) )

# first plot
p1 + geom_point()

# second plot: points connected with lines
p1 + geom_line()
````

````r
p2 <- ggplot(mpg, aes(displ) )

# first plot
p2 + geom_histogram(col="navy",fill="violet")

# second plot
p2 + geom_histogram(col='navy',fill='orchid2',binwidth=0.3)
````

### **Functionals**

DEF: A **functional** is a function that takes functions as input & returns a vector as output. Let us show some examples:

- `map(vec, f)`:   returns`list( f(vec[1]), f(vec[2], ...)`

````r
squareme <- function(x) x^2

unlist(   purrr::map(1:5, squareme)  )
````

````r
lst <- list( a1 = 1:10, b1 = runif(10), c1 = 2:11)

unlist(   purrr::map(lst, mean)  )
````

Any additional arguments are passed to f at each call.

````r
square2 <- function(x, na_replace = 2) {
    if(is.na(x)) return(na_replace^2)
    else x^2
}

vec <- c(0,2,3,NA,7,8)
unlist(  purrr::map(vec, square2, na_replace = 10)   )
````

- `map_lgl()`, `map_int()`, `map_dbl()` and `map_chr()` work in the same way, but they are enforced to return a vector of specific type (logical, int, double or char, respectively)
- `map_dfr()` and `map_dfc()` follow the same fashion, but return a dataframe by rows/columns

````r
# OBS: you can do fancy things!
1:10 %>% map(rnorm, n=20) %>% map_dbl(mean)

# equiv:   map_dbl( map(1:10, rnorm, n=20) , mean)
````
